## Import libraries and number_authors .csv file

```python
#Python notebook by Amedeo Baragiola

import pandas as pd
from matplotlib import pyplot as plt
from plotly.offline import init_notebook_mode, iplot
from sklearn.model_selection import train_test_split
import seaborn as sns

#number_authors.txt is the output of the number_authors job
task1_csv = pd.read_csv('number_authors.txt')

#label columns
task1_csv.columns=['BIN', 'PUBLICATIONS']
```

## Plot number of authors (in bins) and number of publications

```python
plt.figure(figsize=(12, 8))

#plot data
splot = sns.barplot(data=task1_csv, x = 'BIN', y = 'PUBLICATIONS', ci = None)
plt.suptitle('Publications with number of authors inside bin (NumberAuthors)', fontsize=14)
plt.xticks(rotation=45)

#create annotations (put labels on top of columns)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```

## Import author_venue file

```python
#author_venue.txt is the output of the author_venue job
task2_csv = pd.read_csv('author_venue.txt')

#label columns
task2_csv.columns=['BIN', 'TYPE', 'YEAR', 'PUBLICATIONS']
```

## Plot publications w.r.t number of authors (in bins) and venue

```python
sns.set()

#plot data
ax = task2_csv.groupby(['BIN','TYPE']).size().unstack().plot(kind='bar',stacked=True, figsize=(12, 8))
plt.suptitle('Publications by number of author and venue (AuthorVenue)', fontsize=14)
plt.show()
```

## Plot publications w.r.t number of authors (in bins) and year

```python
sns.set()

#plot data
ax = task2_csv.groupby(['BIN','YEAR']).size().unstack().plot(kind='bar',stacked=True, figsize=(12, 8))
plt.suptitle('Publications by number of author and year (AuthorVenue)', fontsize=14)
plt.show()
```
